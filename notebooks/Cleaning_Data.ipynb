```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pbtype = pd.read_csv("pbtype_2015_2021.csv")
```

```python
#remove 2015
print("Before",pbtype.shape)
pbtype = pbtype[pbtype.YEAR != 2015]
print("After",pbtype.shape)
```

```python
#remove rows that have a PBPTYPENAME != Bicyclist

print("Before",pbtype.shape)
pbtype = pbtype[pbtype.PBPTYPENAME == "Bicyclist"]
print("After",pbtype.shape)
```

```python
#make list of columns to remove

removecols = ["VEH_NO","PEDCTYPE", "PEDCTYPENAME", "PEDLOC",
              "PEDLOCNAME","PEDPOS","PEDPOSNAME", "PEDDIR",
              "PEDDIRNAME", "MOTDIR","MOTDIRNAME","MOTMAN",
              "MOTMANNAME","PEDLEG","PEDLEGNAME","PEDSNR",
              "PEDSNRNAME","PEDCGP","PEDCGPNAME"]

print("Before removing", len(removecols),"columns:",pbtype.shape)

for col in removecols:
    pbtype = pbtype.drop(col, axis=1)
    
print("After removing", len(removecols),"columns:",pbtype.shape)
```

```python
pbtype.head(10)
```

```python
print("Before removing unnamed col:",pbtype.shape)
pbtype = pbtype.drop("Unnamed: 0", axis=1)
print("After removing unnamed col:",pbtype.shape)
```

```python
pbtype.head(10)
```

```python
print("Before:",pbtype.columns)
pbtype = pbtype.reset_index(drop=True)
print("After:",pbtype.columns)
```
