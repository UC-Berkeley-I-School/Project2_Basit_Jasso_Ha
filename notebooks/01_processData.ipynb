```python
import pandas as pd
from pathlib import Path
```

```python
ROOT = Path('/home/qha/github/Project2_Bassit_Jasso_Ha')
RAW = ROOT / 'data' / 'raw'
PROCESSED = ROOT / 'data' / 'processed'
```

```python
YEARS = [i for i in range(2014, 2022)]
```

# Import and stack pbtype data

```python
def import_csv(path, name, years=YEARS):
    out_lst = []
    cols_lst = []
    for year in years:
        new_path = RAW / str(year)
        df = pd.read_csv(new_path / (name +'.csv'), encoding='cp1252')
        cols = pd.DataFrame(list(df.columns), columns=['VARNAME'])
        cols['YEAR'] = year
        df['YEAR'] = year
        cols_lst.append(cols)
        out_lst.append(df)
    out_df = pd.concat(out_lst)
    cols_df = pd.concat(cols_lst)
    cols_df = cols_df.pivot_table(columns='YEAR', values='VARNAME', index=cols_df.index, aggfunc='first')    
    return (out_df, cols_df)
```

```python
(stacked_pbtype, cols) = import_csv(PROCESSED, name='pbtype')
cols
```

```python
# remove 2014 from the sample since the set of fields had changed since then
stacked_pbtype = stacked_pbtype[stacked_pbtype['YEAR']>2014]
```

```python
stacked_pbtype.to_csv(PROCESSED / 'pbtype_2015_2021.csv', encoding='utf-8')
```
